```python
import numpy as np
```

```python
def generateTable(data,k=4):
    T = {}
    for i in range(len(data)-k):
        X = data[i:i+k]
        Y = data[i+k]
        #print("X is %s and Y is %s "%(X,Y))
        if T.get(X) is None:
            T[X] = {}
            T[X][Y] = 1
        else:
            if T[X].get(Y) is None:
                T[X][Y] = 1
            else:
                T[X][Y] += 1
        
    for ix in T.keys():
        s = float(sum(T[ix].values()))
        for k in T[ix].keys():
            T[ix][k] = T[ix][k]/s
            
    return T
```

### Using Markov Chain to generate text

```python
text_path = "mypoetry.txt"
def generate_text(fileName):
    with open(fileName) as f:
        return f.read().lower()
```

```python
text = generate_text(text_path)
print(text[:1003])
```

```python
model = generateTable(text)
```

```python
print(model)
```

```python
def sample_next(ctx,T,k=4):
    ctx = ctx[-k:]
    if T.get(ctx) is None:
        return " "
    possible_chars = list(T[ctx].keys())
    possible_values = list(T[ctx].values())
    
    #print(possible_chars)
    #print(possible_values)
    
    return np.random.choice(possible_chars,p=possible_values)
```

```python
def generateText(starting_sent,k=4,maxLen=1000):
    sentence = starting_sent
    ctx = starting_sent[-k:]
    
    for ix in range(maxLen):
        next_prediction = sample_next(ctx,model)
        sentence += next_prediction
        ctx = sentence[-k:]
        
    return sentence
```
